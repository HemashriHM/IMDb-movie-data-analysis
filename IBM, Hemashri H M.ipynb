# 22.Find unique values from genre

```python
data.columns
```

```python
data['Genre']
```

```python
#add , after every elemnt  value.split()
#To create a single list    
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
list1
```

```python
#Convert 2d list to 1d list
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
uni_list
```

```python
len(uni_list)
```
